```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
artists = pd.read_csv('artists-data.csv')
artists.head()
```

```python
songs = pd.read_csv('lyrics-data.csv')
songs.head()
```

```python
artists['Genres'].unique()
```

```python
from itertools import chain
```

```python
genres_list = [str(s).split(';') for s in artists['Genres'].unique()] # Separação dos generos em lista
res = list(chain(*genres_list)) # junção de sublistas
res = [gen.strip() for gen in res] # Tira os espaços antes e depois do genero
genres = list(set(res)) # pega os valores unicos
print(genres)
```

```python
import re
```

```python
for genre in genres:
    contains = [True if re.search(genre, str(art_gen)) else False for art_gen in artists['Genres']]
    artists[genre] = contains
```

```python
artists.head()
```

filtrar musicas country

```python
country_artists = artists[artists['Country']==True]['Link'].unique()
```

Escolhendo somente musicas em ingles

```python
en_songs = songs[songs['language']=='en']
```

```python
all_songs = en_songs.merge(artists, how='outer', left_on='ALink', right_on='Link')
```

```python
all_songs
```

```python
no_country = all_songs[all_songs['Country']==False].sample(n=2500, random_state=1)
yes_country = all_songs[all_songs['Country']==True].sample(n=2500, random_state=1)
```

```python
all_songs = pd.concat([no_country,yes_country]).reset_index()
```

```python
all_songs.head()
```

```python
all_songs['Country'].unique()
```

```python
len(all_songs['Country'])
```

```python
lyrics = all_songs['Lyric'].astype(str)
```

```python
low = [lyr.lower() for lyr in lyrics]
```

```python
tokenized = [word_tokenize(lyr) for lyr in low]
print(tokenized[0])
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
stop_vec = []
for tok in tokenized:
    filtered_stop = [w for w in tok if w not in stop_words]
    stop_vec.append(filtered_stop)
```

```python
stop_vec[0]
```

```python
clean_vec = []
for lyr in stop_vec:
    clean_lyr = [word for word in lyr if word.isalpha()]
    clean_vec.append(clean_lyr)
```

```python
clean_vec[0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
lyrics = []
for lyr in clean_vec:
    lyr_text = ' '.join(lyr)
    lyrics.append(lyr_text)
vectorize = TfidfVectorizer(min_df=5,max_df=0.8)
vectors = vectorize.fit_transform(lyrics)
feature_names = vectorize.get_feature_names_out()
dense = vectors.todense()
dense_list = dense.tolist()
df = pd.DataFrame(dense_list, columns=feature_names)
```

```python
df.head()
```

```python
all_songs['Country'].value_counts()
```

```python
df['Country'] = all_songs['Country'].copy()
```

```python
df['Country'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df['Country'].copy()
x = df.drop('Country', axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
x_train
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
clf = MultinomialNB(alpha=0.1)
```

```python
clf.fit(x_train, y_train.astype(bool))
```

```python
y_pred = clf.predict(x_test)
```

```python
from sklearn import metrics
```

```python
score = metrics.accuracy_score(y_test.astype(bool),y_pred)
score
```

```python
precision = metrics.precision_score(y_test.astype(bool), y_pred)
precision
```

```python
print(metrics.confusion_matrix(y_test.astype(bool), y_pred))
```

```python
recall = metrics.recall_score(y_test.astype(bool), y_pred)
recall
```
